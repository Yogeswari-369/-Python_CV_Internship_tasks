```python
# 1.•	Write a program to loop through a list of numbers and add +2 to every value to elements in list

n=int(input("Enter the n value"))
l=[]
for i in range(0,n):
    l.append(i)
print("Original List is:   ",l)
a=[]
for e in l:
    e+=2
    a.append(e)
print("Add 2 for every element:   ",a)
```

```python
# 3.Program to Print the Fibonacci sequence

def fib(n):
    if n <=1:
        return n
    else:
        return fib(n-1) + fib(n-2)
n = int(input("Enter a number: "))     
for i in range(n):
    print(fib(i),end=" ")
```

```python
# 7.convert the number of days to ages

n= int(input("Enter number of days: "))
years = n// 365
months=(n- years*365)//30
days = (n-years* 365 - months*30)
print("Years = ", years)
print("Months = ", months)
print("Days = ", days)
```

```python
# 9. •	Create a calculator only on a code level by using if condition (Basic arithmetic calculation)


def add(num1, num2):
    return num1 + num2
def subtract(num1, num2):
    return num1 - num2
def multiply(num1, num2):
    return num1 * num2
def divide(num1, num2):
    return num1 / num2
  
print("select operation -\n" \
        "1. Add\n" \
        "2. Subtract\n" \
        "3. Multiply\n" \
        "4. Divide\n")
n1= int(input("Enter first number: "))
n2= int(input("Enter second number: "))
select = int(input("Select operations 1, 2, 3, 4 :"))
if select == 1:
    print(n1, "+", n2, "=",add(n1,n2))
elif select == 2:
    print(n1, "-", n2, "=",subtract(n1,n2))
elif select == 3:
    print(n1, "*", n2, "=",multiply(n1,n2))
elif select == 4:
    print(n1, "/", n2, "=",divide(n1,n2))
else:
    print("Invalid input")
```
