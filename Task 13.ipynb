```python
# 1. Write a Python program for all the cases which can check a string contains only a certain set of characters (in this case a-z, A-Z and 0-9).

import re
def check(str):
    a = re.compile(r'[^a-zA-Z0-9.]')
    str = a.search(str)
    return not bool(str)

print(check("ABCDE123450"))
print(check("abcABC123"))
print(check("*&%@#!}{"))
```

```python
# 3.Write a Python program to check for a number at the end of a word/sentence.

import re
def check(str):
    a= re.compile(r".*[0-9]$")
    if a.match(str):
        return ("True")
    else:
        return ("False")

print(check('abcdef6'))
print(check('abcdef'))
print(check("abcdef1"))
```

```python
# 4.Write a Python program to search the numbers (0-9) of length between 1 to 3 in a given string

import re
ans = re.finditer(r"([0-9]{1,3})", "Exercises number 1, 22, 333,44444 ,55431345 are important")
print("Number of length 1 to 3")
for n in ans:
     print(n.group(0))
```
