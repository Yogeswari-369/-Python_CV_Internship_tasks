  # 2. Design a simple calculator app with try and except for all use cases

```python
def cal():
    try:
        print('+')
        print('-')
        print('*')
        print('%')
        print('/')
        print('**')
        sel=input("Select an operator")
        n1=int(input("Enter n1"))
        n2=int(input("Enter n2"))
        if(sel=='+'):
            print(n1+n2)
        elif(sel=='-'):
            print(n1-n2)
        elif(sel=='*'):
            print(n1*n2)
        elif(sel=='%'):
            print(n1%n2)
        elif(sel=='/'):
            print(n1/n2)
        elif(sel=='**'):
            print(n1**n2)
        else:
            print("Invalid Selection")
    except Exception as a:
        print(a)
        
```

```python
cal()
```
